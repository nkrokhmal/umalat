```python
import pandas as pd
```

## Read csv

```python
path_new = r'/home/nkrokhmal/Downloads/ОперативныеОстатки_20211230.csv'
df_new = pd.read_csv(path_new, sep=';')
df_new.head()
```

## Update mozzarella

```python
path = r'/home/nkrokhmal/Desktop/umalat/app/data/static/params/mozzarella.xlsx'
```

```python
df = pd.read_excel(path)
df.head()
```

```python
for i in range(df_new.shape[0]):
    code = df_new.iloc[0]['Код']
    name = df_new.iloc[0]['Наименование']
    df.loc[df['Kод'] == code, 'Название SKU'] = name
```

```python
df = df[df['Название SKU'].isin(df_new['Наименование'])]
```

```python
df.to_excel(path)
```
